```python
import seaborn as sns
import pandas as pd
```

```python
from sklearn import datasets
```

```python
df=pd.read_csv("Iris.csv")
```

```python
print(df.shape)

print(df.describe())

print(df.dtypes)
```

```python
sns.scatterplot(data=df,x=df['SepalLengthCm'],y=df['SepalWidthCm'],hue=df['Species'],legend="full")
```

```python
sns.lineplot(x=df['SepalLengthCm'],
             y=df['SepalWidthCm'],
             hue=df['Species'],
             data=df)
```

```python
sns.boxplot(x=df['PetalLengthCm'], y=df['PetalWidthCm'], data=df, hue=df['Species'])
```

```python
co_mtx = df.corr(numeric_only=True)
print(co_mtx)
sns.heatmap(co_mtx, cmap="YlGnBu", annot=True)
```

```python
sns.catplot(x=df['PetalLengthCm'],hue=df['Species'],
                data=df)
```

```python
df1=pd.read_csv("Titanic-dataset.csv")
```

```python

print(df1.shape)

print(df1.describe())

print(df1.dtypes)
```

```python
sns.scatterplot(data=df1,x=df1['Age'],y=df1['Fare'],hue=df1['Survived'])
```

```python
correlation = df1[['Age', 'Fare']].corr()
print(correlation)
```

```python
sns.lineplot(x=df1['Age'],
             y=df1['Fare'],
             hue=df1['Survived'],
             data=df1)
```

```python
sns.boxplot(x=df1['Pclass'], y=df1['SibSp'], data=df1, hue=df1['Survived'])
```

```python
co_mtx = df1.corr(numeric_only=True)
print(co_mtx)
sns.heatmap(co_mtx, cmap="YlGnBu", annot=True)
```
