```python
def fun(value):
    if value > 0:
        return value
    else:
        return df1['age'].median()
def fun_gender(value):
    if value == 'Male':
        return 'M'
    elif value == 'Female':
        return 'F'
    else:
        return value
def fun_income(value):
    if value <= 0:
        return df1['income'].mean()
    else:
        return value
df1['age'] = df1['age'].apply(fun)
df1['gender'] = df1['gender'].apply(fun_gender)
df1['name']=df1['name'].fillna('Nan')
df1[df1['age']>0]=df1[df1['age']>0].fillna(0)
df1['income'] = df1['income'].apply(fun_income)
df1['joined_date'] = pd.to_datetime(df1['joined_date'],format='mixed')
```
