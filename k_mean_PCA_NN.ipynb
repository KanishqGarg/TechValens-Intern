```python
import pandas as pd
from sklearn.cluster import KMeans,AgglomerativeClustering
import seaborn as sns
import matplotlib.pyplot as plt
df=pd.read_csv('Mall_customers.csv')
```

```python
df.describe()
df.drop(['Genre'], axis=1, inplace=True)
```

```python
kmeans = KMeans(n_clusters=3, random_state=0, n_init="auto").fit(df)
kmeans.labels_
df1=df
df1['label']=kmeans.predict(df)
```

```python
sns.scatterplot(x=df1['Spending Score (1-100)'],y=df1['Annual Income (k$)'],hue=df1['label'],data=df1)
```

```python
sns.scatterplot(x=df1['Annual Income (k$)'],y=df1['Age'],hue=df1['label'],data=df1)
```

```python
heir = AgglomerativeClustering(n_clusters=3).fit(df)
heir.labels_
df2=df
df2['label']=heir.fit_predict(df)
```

```python
sns.scatterplot(x=df2['Spending Score (1-100)'],y=df2['Annual Income (k$)'],hue=df2['label'],data=df2)
```

```python
sns.scatterplot(x=df2['Annual Income (k$)'],y=df2['Age'],hue=df2['label'],data=df2)
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
df=scaler.fit_transform(df)
pca = PCA(n_components=2)
df_pca=pca.fit_transform(df)
```

```python
print(df_pca[:0],df_pca[:1])
# sns.scatterplot(x=df_pca[:0],y=df_pca[:1])
```

```python
plt.figure(figsize=(8,6))
plt.scatter(df_pca[:, 0], df_pca[:, 1], edgecolor='k')
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.title("PCA Transformed Data")
plt.show()
```
